```python
import pandas as pd
import matplotlib.pyplot as plt

# 讀取檔案
orgdata = pd.read_csv("澎湖.csv")

# 選擇所需的欄位
columns_to_compare = ["年分-110年：航空人次", "年分-111年：航空人次",
                      "年分-110年：輪船人次", "年分-111年：輪船人次",
                      "年分-110年：觀光客人次", "年分-111年：觀光客人次",
                      "年分-111年：觀光客增減人數"]

# 比較110年和111年的各項目
comparison = orgdata[columns_to_compare]

# 繪製長條圖
comparison.plot(kind='bar', figsize=(10, 6))
plt.xlabel('項目')
plt.ylabel('人次')
plt.title('110年與111年各項目比較')
plt.legend(['110年', '111年'])
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 讀取檔案並指定數據類型
orgdata = pd.read_csv("澎湖.csv")

# 選擇所需的欄位
columns = ["月份", "年分-111年：觀光客增減人數", "年分-111年：觀光客人次"]

# 提取需要的資料
data = orgdata[columns]

# 將月份設定為索引
data.set_index("月份", inplace=True)

# 將缺失值或非數字值替換為0
data.replace('-', 0, inplace=True)
data = data.astype(float)

# 繪製熱力圖
plt.figure(figsize=(10, 6))
sns.heatmap(data, annot=True, cmap="YlGnBu", fmt=".0f")
plt.xlabel("指標")
plt.ylabel("月份")
plt.title("年分-111年：觀光客增減人數和年分-111年：觀光客人次熱力圖")

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 讀取檔案
orgdata = pd.read_csv("澎湖.csv")

# 選擇所需的欄位
columns = ["月份", "年分-111年：輪船人次"]

# 提取相應的數據
data = orgdata[columns]

# 進行資料透視
pivot_data = data.pivot_table(index="月份", columns="月份", values="年分-111年：輪船人次")

# 繪製熱力圖
plt.figure(figsize=(8, 6))
sns.heatmap(pivot_data, cmap="YlGnBu", annot=True, fmt=".0f")
plt.xlabel('月份')
plt.ylabel('月份')
plt.title('月份與觀光客人次熱力圖')

plt.show()
```
