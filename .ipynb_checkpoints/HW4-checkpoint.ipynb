```python
import pandas as pd
data = pd.read_csv("data.csv")
```

```python
# the last five data of the list
data.tail()
```

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
danceability = data["danceability"]
tempo = data["tempo"]
```

```python
data_top_20 = data[:20]
data_top_20
```

```python
data_top_20 = data[['danceability', 'tempo']].groupby(['danceability']).sum().sort_values('tempo', ascending=False).head(20)

# 繪製圖表
plt.bar(data_top_20.index, data_top_20['tempo'])
plt.xticks(rotation=90)
plt.xlabel('danceability')
plt.ylabel('tempo')
plt.title('Top 20 danceability by Total tempo')
plt.show()
```
