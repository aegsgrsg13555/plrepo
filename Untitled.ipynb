```python
lolrank_sorted=lolrank.sort_values("blueWardsPlaced")
lolrank_sorted
#將blueWardsPlaced升序排名
```

```python
sum_of_blue_kills = lolrank.loc[lolrank['blueWins'] == 0, 'blueKills'].sum()
print('Sum of blueKills when blueWins = 0:', sum_of_blue_kills)
#將blueWins為0的blueKills相加
```

```python
lolrank_sorted = lolrank.sort_values(by=['blueWardsPlaced'], ascending=False).reset_index(drop=False)
lolrank_sorted.to_csv('lolrank_sorted.csv', index=False)
lolrank_sorted.to_csv('lolrank_sorted.csv', index=True)
print(lolrank_sorted.to_csv)
#自主嘗試將blueWardsPlaced降序排名
```

```python
lolrank = pd.read_csv('lolrank.csv')
# 篩選出blueKills小於5的資料
filtered_lolrank = lolrank[lolrank['blueKills'] < 5]


total_blue_kills = filtered_lolrank['blueKills'].sum()

print('blueKills小於5的總和為:', total_blue_kills)
# 將blueKills小於5的資料進行加總
```

```python
lolrank = pd.read_csv('lolrank.csv')

# 篩選出blueWardsPlaced大於100的資料
filtered_lolrank = lolrank[lolrank['blueWardsPlaced'] > 100]

# 將blueWardsPlaced大於100的資料進行加總
total_blue_wards_placed = filtered_lolrank['blueWardsPlaced'].sum()

print('blueWardsPlaced大於100的總和為:', total_blue_wards_placed)

```
